### Aggregate Data with the Python .agg() Function

Let's work out the number of different themes shipped by year. This means we have to count the number of unique theme_ids per calendar year.

```python
sets_df.groupby('year')['theme_id'].nunique()
```

```python
sets_df.groupby('year').agg({'theme_id':'nunique'}) #group by year and then apply aggregate functionalities on specific columns of the groupby object
```

**Challenge**: Plot the number of themes released by year on a line chart. Only include the full calendar years (i.e., exclude 2020 and 2021).

```python
import matplotlib.pyplot as plt
x=sets_df.groupby('year').agg({'theme_id':'nunique'}).index[:-2]
y=sets_df.groupby('year').agg({'theme_id':'nunique'}).theme_id[:-2]
plt.plot(x,y)
```

### Line Charts with Two Seperate Axes

```python
sets_by_year=sets_df.groupby('year').count()
themes_by_year=sets_df.groupby('year').agg({'theme_id':'nunique'})
themes_by_year.head()
```

```python
plt.plot(sets_by_year.index[:-2],sets_by_year.set_num[:-2])
plt.plot(themes_by_year.index[:-2],themes_by_year.theme_id[:-2])
#not that clear
```

```python
ax1=plt.gca()#get current axes
ax2=ax1.twinx()#make a second y-axis sharing the same x-axis

ax1.plot(sets_by_year.index[:-2],sets_by_year.set_num[:-2],'g')
ax2.plot(themes_by_year.index[:-2],themes_by_year.theme_id[:-2],'r')
```

**Challenge**: Use the <code>.groupby()</code> and <code>.agg()</code> function together to figure out the average number of parts per set. How many parts did the average LEGO set released in 1954 compared to say, 2017?

```python
parts_per_set=sets_df.groupby('year').agg({'num_parts':'mean'})
parts_per_set.head()
```

### Number of Sets per LEGO Theme

LEGO has licensed many hit franchises from Harry Potter to Marvel Super Heros to many others. But which theme has the largest number of individual sets?

```python
sets_theme_count=sets_df.groupby('theme_id').agg({'set_num':'count'}).sort_values(by='set_num',ascending=False)
sets_theme_count.head()
```

### Merging (i.e., Combining) DataFrames based on a Key


```python
sets_theme_count_df=pd.DataFrame({'id':sets_theme_count.index,
                                  'set_count':sets_theme_count['set_num'].values}) #Gives the list of values
#You should always pass a 1D list/array as value
```

```python
merged_df=pd.merge(sets_theme_count_df,themes_df,on='id')
merged_df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,8))
plt.xticks(fontsize=14,rotation=45)
plt.yticks(fontsize=14)
plt.xlabel('Theme name', fontsize=14)
plt.ylabel('Number of sets',fontsize=14)
plt.bar(merged_df.name[:10],merged_df.set_count[:10])
```
