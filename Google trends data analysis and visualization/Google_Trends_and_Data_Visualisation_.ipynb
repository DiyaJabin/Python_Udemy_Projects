### Converting from Daily to Monthly Data

### Tesla Stock Price v.s. Search Volume

**Challenge:** Plot the Tesla stock price against the Tesla search volume using a line chart and two different axes. Label one axis 'TSLA Stock Price' and the other 'Search Trend'.

```python
import matplotlib.pyplot as plt

ax1=plt.gca()
ax2=ax1.twinx()

ax1.plot(df_tesla.MONTH,df_tesla.TSLA_USD_CLOSE,color="skyblue")
ax2.plot(df_tesla.MONTH,df_tesla.TSLA_WEB_SEARCH,color="#B8A992")
ax1.set_ylabel('TSLA Stock Price',color="skyblue")
ax2.set_ylabel('Search Trend',color="#B8A992")
```

### Unemployement Benefits Search vs. Actual Unemployment in the U.S.

**Challenge** Plot the search for "unemployment benefits" against the unemployment rate.
1. Change the title to: Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate <br>

```python
plt.figure(figsize=(14,8),dpi=120) #width 24 inches, height 8 inches and resolution 120 dots per inch
plt.title('Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate',fontsize=18)

#set the x-axis tick label fontsize to 14 and rotate labels by 45
plt.xticks(fontsize=14,rotation=45)

ax1=plt.gca()
ax2=ax1.twinx()
#set y-axis range
ax1.set_ylim([3,10.5])
 #set x-axis range
ax1.set_xlim([df_unemployment.MONTH.min(),df_unemployment.MONTH.max()])
ax1.grid(color='grey',linestyle='--')

ax1.plot(df_unemployment.MONTH,df_unemployment.UNRATE,color="skyblue",linestyle='--')
ax2.plot(df_unemployment.MONTH,df_unemployment.UE_BENEFITS_WEB_SEARCH,marker='o')
ax1.set_ylabel('FRED U/E Rate',color="skyblue",fontsize=14)
ax2.set_ylabel('Search Trend',color="#B8A992",fontsize=14)
```

**Challenge**: Calculate the 3-month or 6-month rolling average for the web searches. Plot the 6-month rolling average search data against the actual unemployment. What do you see in the chart? Which line moves first?


```python
df_unemployment['UE_BENEFITS_WEB_SEARCH']=df_unemployment['UE_BENEFITS_WEB_SEARCH'].rolling(window=6).mean()
df_unemployment['UNRATE']=df_unemployment['UNRATE'].rolling(window=6).mean()
plt.figure(figsize=(14,8),dpi=120) #width 24 inches, height 8 inches and resolution 120 dots per inch
plt.title('Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate',fontsize=18)

#set the x-axis tick label fontsize to 14 and rotate labels by 45
plt.xticks(fontsize=14,rotation=45)

ax1=plt.gca()
ax2=ax1.twinx()
#set y-axis range
ax1.set_ylim([3,10])
#set x-axis range
ax1.set_xlim([df_unemployment.MONTH.min(),df_unemployment.MONTH.max()])
ax1.grid(color='grey',linestyle='--')

ax1.plot(df_unemployment.MONTH,df_unemployment.UNRATE,color="skyblue",linestyle='--')
ax2.plot(df_unemployment.MONTH,df_unemployment.UE_BENEFITS_WEB_SEARCH,marker='o')
ax1.set_ylabel('FRED U/E Rate',color="skyblue",fontsize=14)
ax2.set_ylabel('Search Trend',color="#B8A992",fontsize=14)
```
